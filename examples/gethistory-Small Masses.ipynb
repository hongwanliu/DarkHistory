```python
%autoreload

import matplotlib
import matplotlib.pyplot as plt
matplotlib.rc_file('matplotlibrc')

import numpy as np
from scipy.interpolate import interp1d

import darkhistory.physics as phys
import darkhistory.history.tla as tla
```

```python
rs_vec=np.flipud(np.arange(4, 3000., 0.1))
```

```python
soln_default_reion = tla.get_history(rs_vec, helium_TLA=True, reion_switch=True)


soln_default_reion_DM_1  = tla.get_history(
    rs_vec, baseline_f = True,
    inj_particle = 'phot', DM_process = 'decay',
    mDM=5e0, lifetime = 3e23,
    reion_switch=True, helium_TLA=True,
)

soln_default_reion_DM_2  = tla.get_history(
    rs_vec, baseline_f = True,
    inj_particle = 'phot', DM_process = 'decay',
    mDM=5e2, lifetime = 3e23,
    reion_switch=True, helium_TLA=True,
)

soln_default_reion_DM_3  = tla.get_history(
    rs_vec, baseline_f = True,
    inj_particle = 'phot', DM_process = 'decay',
    mDM=5e3, lifetime = 3e23,
    reion_switch=True, helium_TLA=True,
)

soln_default_reion_DM_4  = tla.get_history(
    rs_vec, baseline_f = True,
    inj_particle = 'phot', DM_process = 'decay',
    mDM=1e4, lifetime = 3e23,
    reion_switch=True, helium_TLA=True,
)

soln_default_reion_DM_5  = tla.get_history(
    rs_vec, baseline_f = True,
    inj_particle = 'phot', DM_process = 'decay',
    mDM=1.1e4, lifetime = 3e23,
    reion_switch=True, helium_TLA=True,
)

soln_default_reion_DM_6  = tla.get_history(
    rs_vec, baseline_f = True,
    inj_particle = 'phot', DM_process = 'decay',
    mDM=2e4, lifetime = 3e23,
    reion_switch=True, helium_TLA=True,
)
soln_default_reion_DM_7  = tla.get_history(
    rs_vec, baseline_f = True,
    inj_particle = 'phot', DM_process = 'decay',
    mDM=1e5, lifetime = 3e23,
    reion_switch=True, helium_TLA=True,
)
```

```python
Tm_DM_7 = soln_default_reion_DM_7[:,0]/phys.kB
Tm_DM_6 = soln_default_reion_DM_6[:,0]/phys.kB
Tm_DM_5 = soln_default_reion_DM_5[:,0]/phys.kB
Tm_DM_4 = soln_default_reion_DM_4[:,0]/phys.kB
Tm_DM_3 = soln_default_reion_DM_3[:,0]/phys.kB
Tm_DM_2 = soln_default_reion_DM_2[:,0]/phys.kB
Tm_DM_1 = soln_default_reion_DM_1[:,0]/phys.kB
Tm = soln_default_reion[:,0]/phys.kB

plt.figure(figsize=(15,6.5))

ax = plt.subplot(1,2,1)

plt_TIGM_1, = plt.plot(rs_vec, Tm_DM_1, '--', label=r'$T_m (m_\chi=5eV)$')
#plt_TIGM_2, = plt.plot(rs_vec, Tm_DM_2, ':', label=r'$T_m (m_\chi=0.5keV)$')
plt_TIGM_3, = plt.plot(rs_vec, Tm_DM_3, linestyle='dashdot',  label=r'$T_m (m_\chi=5keV)$')
plt_TIGM_4, = plt.plot(rs_vec, Tm_DM_4,':',  label=r'$T_m (m_\chi=10keV)$')
plt_TIGM_5, = plt.plot(rs_vec, Tm_DM_5, '--', label=r'$T_m (m_\chi=11keV)$')
plt_TIGM_6, = plt.plot(rs_vec, Tm_DM_6, linestyle='dashdot', label=r'$T_m (m_a=20keV)$')
plt_TIGM_7, = plt.plot(rs_vec, Tm_DM_7, '--', label=r'$T_m (m_\chi=100keV)$')
plt_TIGM_noDM, = plt.plot(rs_vec, Tm,  ':', label=r'$T_m$ no DM')

ax.set_xscale('log')
ax.set_yscale('log')
# Text
plt.text(0.65, 0.12, r'$\chi \to \gamma \gamma$', fontsize=20, transform=ax.transAxes) 
plt.text(0.65, 0.05, r'$\tau = 3 \times 10^{23}$ s', fontsize=20, transform=ax.transAxes)

plt.axis([4, 1e1, 7e3, 3e4])
plt.title(r'\bf{get.history, default reionization}')
plt.xlabel('Redshift (1+z)')
plt.ylabel('Temperature [K]')

leg=plt.legend(handles=[ plt_TIGM_1, plt_TIGM_3, plt_TIGM_4, plt_TIGM_5, plt_TIGM_6, plt_TIGM_7, plt_TIGM_noDM], loc=1)

plt.savefig('MatterTemperature-NoBackrection-DifferentMasses.png')
```
